# PNG Thumbnail Generator

This notebook demonstrates how to create a thumbnail image in PNG format using Python and the Pillow library.

```python
from PIL import Image, ImageDraw, ImageFont
from pathlib import Path

print('Pillow imported successfully')
```

## Load an Image

Load an image file from disk using `Image.open`. The notebook will create a sample image if `input.jpg` does not exist.

```python
input_path = Path('input.jpg')
sample_path = Path('sample_input.jpg')
if not input_path.exists():
    print('input.jpg not found, generating a sample image...')
    sample_image = Image.new('RGB', (512, 512), '#4a90e2')
    draw = ImageDraw.Draw(sample_image)
    draw.rectangle([64, 64, 448, 448], outline='white', width=8)
    draw.text((128, 220), 'Sample Image', fill='white')
    sample_image.save(sample_path, format='PNG')
    input_path = sample_path

image = Image.open(input_path)
print('Loaded image:', input_path)
print('Original size:', image.size)
image
```

## Resize the Image to Thumbnail Size

Resize the loaded image to a smaller thumbnail size while preserving aspect ratio.

```python
thumbnail_size = (128, 128)
thumbnail_image = image.copy()
thumbnail_image.thumbnail(thumbnail_size)
print('Thumbnail size:', thumbnail_image.size)
thumbnail_image
```

## Save the Thumbnail as PNG

Save the resized image as `thumbnail.png` in this workspace.

```python
output_path = Path('thumbnail.png')
thumbnail_image.save(output_path, format='PNG')
print('Saved thumbnail as:', output_path)
print('File exists:', output_path.exists())
```
